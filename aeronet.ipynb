```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
aeronet = pd.read_csv("data/aeronet.csv",header=6)
datetime_str = aeronet['Date(dd:mm:yyyy)'] + ' ' + aeronet['Time(hh:mm:ss)']
aeronet['datetime'] = pd.to_datetime(datetime_str, format='%d:%m:%Y %H:%M:%S')
aeronet.set_index('datetime', inplace=True)
aeronet.drop(['Date(dd:mm:yyyy)', 'Time(hh:mm:ss)'], axis=1, inplace=True)

aeronet['Refractive_Index-Real_Part[675nm]'] = aeronet['Refractive_Index-Real_Part[675nm]'].astype(float)
```

```python
real = aeronet['Refractive_Index-Real_Part[675nm]']
imaginaria = aeronet['Refractive_Index-Imaginary_Part[675nm]']
```

## Verão


```python
start_inverno = "2018-06-01 00:00:00"
start_inverno = datetime.strptime(start_inverno, "%Y-%m-%d %H:%M:%S")

end_inverno = "2018-08-31 23:59:59"
end_inverno = datetime.strptime(end_inverno, "%Y-%m-%d %H:%M:%S")

start_primavera= "2017-09-01 00:00:00"
start_primavera = datetime.strptime(start_primavera, "%Y-%m-%d %H:%M:%S")

end_primavera= "2017-11-30 23:59:59"
end_primavera = datetime.strptime(end_primavera , "%Y-%m-%d %H:%M:%S")



start_verao = "2017-12-01 00:00:00"
start_verao = datetime.strptime(start_verao, "%Y-%m-%d %H:%M:%S")

end_verao = "2018-02-28 23:59:59"
end_verao = datetime.strptime(end_verao, "%Y-%m-%d %H:%M:%S")
```

```python

plt.figure(figsize=(12, 6))
plt.scatter(imaginaria.index, real,s=20)
#plt.plot(nefe.index, green, color='forestgreen', label='Total scatt Green')
#plt.plot(nefe.index, blue, color='royalblue', label='Total scatt Blue')

plt.axvspan(pd.to_datetime(start_inverno), pd.to_datetime(end_inverno), color='blue', alpha=0.3, label='Inverno')
plt.axvspan(pd.to_datetime(start_primavera), pd.to_datetime(end_primavera), color='green', alpha=0.5, label='Primavera')
plt.axvspan(pd.to_datetime(start_verao), pd.to_datetime(end_verao), color='orange', alpha=0.5, label='Verão')
# Formatando o eixo X para mostrar a nefe e hora
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y - %H:%M'))
plt.gcf().autofmt_xdate()  # Rotaciona os rótulos do eixo X para melhor legibilidade

plt.legend()  # Mostra a legenda
plt.grid()  # Adiciona a grade ao gráfico
plt.show()
```

```python
print('Inverno')
real_inverno = real.loc[start_inverno:end_inverno]
imaginaria_inverno = imaginaria.loc[start_inverno:end_inverno]
print('n = ',real_inverno.mean())
print('k = ',imaginaria_inverno.mean())
```

```python
print('Primavera')
real_primavera = real.loc[start_primavera:end_primavera]
imaginaria_primavera = imaginaria.loc[start_primavera:end_primavera]
print('n = ',real_primavera.mean())
print('k = ',imaginaria_primavera.mean())
```

```python
print('Verão')
real_verao = real.loc[start_verao:end_verao]
imaginaria_verao = imaginaria.loc[start_verao:end_verao]
print('n = ',real_verao.mean())
print('k = ',imaginaria_verao.mean())
```
